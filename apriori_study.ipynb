# Apriori v1

## Import  e tratamento dos dados

```python
import pandas as pd 
```

```python
data = pd.read_csv('../Projects/FINAL_DATASET_MARKETING.csv')
data = data.rename(columns={'Unnamed: 0': 'TransactionNo', 'produtos': 'Items'})
data.head()
```

```python
#transformando a coluna 'Items' em uma série do tipo string
coluna_items = pd.Series(data['Items'], dtype="string")
coluna_items 
```

```python
import ast

#tranformando a série 'Items'->string em uma lista em que cada posição => info da venda.

lista_itens_unicos = []
for i in range(len(coluna_items)):
    item = ast.literal_eval(coluna_items[i])
    lista_itens_unicos.append(item)
lista_itens_unicos[0:5]
```

## Preparação dos dados

```python
# Aqui transformaremos a lista com as infos de cada venda em um dataframe com as colunas = produtos
```

```python
from mlxtend.preprocessing import TransactionEncoder


te = TransactionEncoder()
lista_itens_unicos_te = te.fit(lista_itens_unicos).transform(lista_itens_unicos)
lista_itens_unicos_te
```

```python
lista_itens_unicos_transformado = pd.DataFrame(lista_itens_unicos_te,columns=te.columns_)
lista_itens_unicos_transformado
```

## Algoritmos & Resultados

```python
#Algoritmo apriori
from mlxtend.frequent_patterns import apriori

items_frequentes_apriori = apriori(lista_itens_unicos_transformado, use_colnames=True, min_support=0.02)
items = items_frequentes_apriori.sort_values(['support'],ascending=False).reset_index()
```

```python
items.head(10) #coleções mais vendidas
```

```python
from mlxtend.frequent_patterns import association_rules
regras_apriori = association_rules(items_frequentes_apriori,metric='confidence',min_threshold=0.50)
regras_apriori.sort_values('confidence',ascending=False) #insights com confiança > 50%
```

```python
# obtendo uma lista com os itens unicos para verificar amostragem e substituir por algum grupo se a distribuicao permitir
items_unicos = []
for i in range(len(lista_itens_unicos)):
    for j in range(len(lista_itens_unicos[i])):
        item_especifico = lista_itens_unicos[i][j]
        if item_especifico not in items_unicos:
            items_unicos.append(item_especifico)

sorted(items_unicos)
    
```

```python
from collections import OrderedDict

distribuicao_itens = {}

def pegar_amostragem(data:,items_unicos:list,annot=False):
    distribuicao_itens = {} #a amostragem será organizada por um dicionário com {'key'-> item :'value'-> frequencia}

    for i in range(len(items_unicos)): #percorrendo todos os itens unicos
        count = 0 #zerando a contagem
        if annot==True:
            print(f'Analisando: {items_unicos[i]}') 
        for j in range(len(data)):
            if items_unicos[i] in data.loc[j]['Items']: #se o item único for encontrado em alguma venda (linha do dataset)
                count = count + 1 #contar as vezes
                if annot==True:
                    print(f'Valor encontrado em {data.loc[j]}! contagem: {count}')
        distribuicao_itens[items_unicos[i]] = count #então, distribuicao_itens['item_especifico'] = frequencia do item nos dados
        
    return  OrderedDict(sorted(distribuicao_itens.items())) #retornar as distribuicoes com as chaves ordenadas alfabeticamente 
```

```python
amostragem = pegar_amostragem(data=data,items_unicos=items_unicos,annot=False)
```
